```python
import numpy as np
import pandas as pd
```

```python
class knn_classifier:
    def __init__(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def fit_predict(self, X_test, k=1,  distance='hamming'):
        pred = []
        for i in range(len(X_test)):
            dists = self.compute_distance(X_test[i], distance)
            label = self.label_predict(dists, k)
            pred.append(label)
        return np.array(pred)
    
    def compute_distance(self, X_test, distance='hamming'):
        if distance == "euclidean":
            dists = np.sqrt(np.sum((self.X_train - X_test)**2, axis=1))

        elif distance == "hamming":
            dists = np.sum(self.X_train != X_test, axis=1)

        elif distance == "manhattan":
            dists = np.sum(np.abs(self.X_train - X_test), axis=1)

        else: 
            raise ValueError("Invalid distance metric") 
        
        return dists # dists is a numpy array (1D)
    
    def label_predict(self, dists, k=1):
        closest_idx = np.argsort(dists)[:k]
        closest_y = self.y_train[closest_idx]

        counts = {}
        for label in closest_y:
            counts[label] = counts.get(label,0)+1

        return max(counts, key=counts.get)
            
        
```
