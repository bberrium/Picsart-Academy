# Pandas Demo: Core Ideas, Not Syntax

This notebook demonstrates:
- Series vs DataFrame
- Indexing and alignment
- Inspecting and questioning data
- GroupBy and summarization
- Transformations and scaling
- Numericalization (one-hot encoding)

This is **not** an API tour.


```python
import pandas as pd
import numpy as np
```

```python
data = {
    "name": ["Hrach", "Ani", "Meri", "Hayk", "Arevik"],
    "city": ["Yerevan", "Yerevan", "Gyumri", "Gyumri", "Yerevan"],
    "age": [23, 29, 35, None, 41],
    "salary": [300_000, 450_000, None, 280_000, 500_000],
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
ages = df["age"]
type(ages)
```

```python
type(df)
```

```python
ages.mean()
```

```python
# Indexing: Labels vs Positions

df.index
```

```python
df.loc[0]
```

```python
df.iloc[0]
```

```python
df = df.set_index("name")
df
```

```python
# alignment in action
salary = df["salary"]
age = df["age"]

salary / age
```

```python
df[["age", "salary"]] - df[["age", "salary"]].mean()
```

Using GroupBy is like asking a question

```python
df.groupby("city")["salary"].mean() # how much is the mean salary by city?
```

```python
df.groupby("city").describe()
```

Creating Meaning With Transformations

```python
df["salary_k"] = df["salary"] / 1000
df
```

```python
df["age_group"] = pd.cut(df["age"], bins=[0, 30, 50], labels=["young", "mid"])
df
```

```python
# scaling
salary_mean = df["salary"].mean()
salary_std = df["salary"].std()

df["salary_z"] = (df["salary"] - salary_mean) / salary_std
df[["salary", "salary_z"]]
```

```python
df[df['city'] == 'Yerevan']
```

```python
df[df['salary'] > 300000]
```

Pandas allows numericalization

```python
df[["city", "age_group"]]
```

```python
# one-hot encoding
pd.get_dummies(df[["city", "age_group"]])
```

```python
encoded = pd.get_dummies(df[["city", "age_group"]])
df_model = pd.concat([df, encoded], axis=1)
df_model
```

Handling missing data

```python
df_model.isna().mean()
```

```python
df_model_filled = df_model.copy()
df_model_filled["age"] = df_model_filled["age"].fillna(df_model_filled["age"].mean())
df_model_filled
```

```python
df_model_filled.describe()
```

## Takeaways

- Pandas works with **labeled data**
- Indexing and alignment drive behavior
- GroupBy is about questions, not syntax
- Transformations encode assumptions
- Numericalization is a design choice

Clean code  correct analysis.
